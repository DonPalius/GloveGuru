```python
# Ensure numeric values for weight, height, and reach are clean and computable
import numpy as np

# Convert height to inches (cleaning the format "5'11" and such)
def height_to_inches(height):
    if height == '--' or pd.isnull(height):
        return np.nan
    try:
        feet, inches = height.split("'")
        inches = inches.replace('"', '').strip()
        return int(feet) * 12 + int(inches)
    except:
        return np.nan

# Convert weight to pounds (removing "lbs.")
def weight_to_pounds(weight):
    if weight == '--' or pd.isnull(weight):
        return np.nan
    try:
        return int(weight.replace('lbs.', '').strip())
    except:
        return np.nan

# Convert reach to inches (removing quote marks and cleaning)
def reach_to_inches(reach):
    if reach == '--' or pd.isnull(reach):
        return np.nan
    try:
        return int(reach.replace('"', '').strip())
    except:
        return np.nan

# Apply the conversions
df_ufc_fighter_tott['HEIGHT_INCHES'] = df_ufc_fighter_tott['HEIGHT'].apply(height_to_inches)
df_ufc_fighter_tott['WEIGHT_LBS'] = df_ufc_fighter_tott['WEIGHT'].apply(weight_to_pounds)
df_ufc_fighter_tott['REACH_INCHES'] = df_ufc_fighter_tott['REACH'].apply(reach_to_inches)

# Calculate the reach-to-weight-to-height ratio
df_ufc_fighter_tott['REACH_WEIGHT_HEIGHT_RATIO'] = (
    df_ufc_fighter_tott['REACH_INCHES'] / (df_ufc_fighter_tott['WEIGHT_LBS'] * df_ufc_fighter_tott['HEIGHT_INCHES'])
)

# Sort the DataFrame by the ratio in descending order
sorted_fighters = df_ufc_fighter_tott.sort_values(by='REACH_WEIGHT_HEIGHT_RATIO', ascending=False)

sorted_fighters.head(25)
```

```python
# Step 1: Split BOUT into Fighter A and Fighter B
df_ufc_fight_details[['FIGHTER_A', 'FIGHTER_B']] = df_ufc_fight_details['BOUT'].str.split(' vs. ', expand=True)

# Step 2: Merge fighter attributes for Fighter A
df_merged = df_ufc_fight_details.merge(
    df_ufc_fighter_tott,
    left_on='FIGHTER_A',
    right_on='FIGHTER',
    how='left'
)

# Rename Fighter A columns
df_merged = df_merged.rename(columns={
    'HEIGHT': 'FIGHTER_A_HEIGHT', 'WEIGHT': 'FIGHTER_A_WEIGHT', 
    'REACH': 'FIGHTER_A_REACH', 'STANCE': 'FIGHTER_A_STANCE'
})

# Drop duplicate 'FIGHTER' column
df_merged = df_merged.drop(columns=['FIGHTER'])

# Step 3: Merge fighter attributes for Fighter B
df_merged = df_merged.merge(
    df_ufc_fighter_tott,
    left_on='FIGHTER_B',
    right_on='FIGHTER',
    how='left'
)

# Rename Fighter B columns
df_merged = df_merged.rename(columns={
    'HEIGHT': 'FIGHTER_B_HEIGHT', 'WEIGHT': 'FIGHTER_B_WEIGHT', 
    'REACH': 'FIGHTER_B_REACH', 'STANCE': 'FIGHTER_B_STANCE'
})

# Drop duplicate 'FIGHTER' column
df_merged = df_merged.drop(columns=['FIGHTER'])
df_merged
```

```python
import numpy as np

# Function to convert height to inches
def height_to_inches(height):
    if height == '--' or pd.isnull(height):
        return np.nan
    try:
        feet, inches = height.split("'")
        inches = inches.replace('"', '').strip()
        return int(feet) * 12 + int(inches)
    except:
        return np.nan

# Convert height, weight, and reach to numerical values
df_merged['FIGHTER_A_HEIGHT'] = df_merged['FIGHTER_A_HEIGHT'].apply(height_to_inches)
df_merged['FIGHTER_B_HEIGHT'] = df_merged['FIGHTER_B_HEIGHT'].apply(height_to_inches)

# Function to convert weight to numerical (lbs.)
def weight_to_pounds(weight):
    if weight == '--' or pd.isnull(weight):
        return np.nan
    try:
        return int(weight.replace('lbs.', '').strip())
    except:
        return np.nan

df_merged['FIGHTER_A_WEIGHT'] = df_merged['FIGHTER_A_WEIGHT'].apply(weight_to_pounds)
df_merged['FIGHTER_B_WEIGHT'] = df_merged['FIGHTER_B_WEIGHT'].apply(weight_to_pounds)

# Function to convert reach to inches
def reach_to_inches(reach):
    if reach == '--' or pd.isnull(reach):
        return np.nan
    try:
        return int(reach.replace('"', '').strip())
    except:
        return np.nan

df_merged['FIGHTER_A_REACH'] = df_merged['FIGHTER_A_REACH'].apply(reach_to_inches)
df_merged['FIGHTER_B_REACH'] = df_merged['FIGHTER_B_REACH'].apply(reach_to_inches)

# Now calculate the differences
df_merged['HEIGHT_DIFF'] = df_merged['FIGHTER_A_HEIGHT'] - df_merged['FIGHTER_B_HEIGHT']
df_merged['REACH_DIFF'] = df_merged['FIGHTER_A_REACH'] - df_merged['FIGHTER_B_REACH']
df_merged['WEIGHT_DIFF'] = df_merged['FIGHTER_A_WEIGHT'] - df_merged['FIGHTER_B_WEIGHT']

# Convert stance to a categorical feature
df_merged['STANCE_MATCHUP'] = (df_merged['FIGHTER_A_STANCE'] == df_merged['FIGHTER_B_STANCE']).astype(int)
df_merged
```

```python
# Merge with fight results to include outcomes, using suffixes to handle duplicates
df_final = df_merged.merge(
    df_ufc_fight_results, 
    on=['EVENT', 'BOUT'], 
    how='left', 
    suffixes=('_merged', '_results')
)

# Now proceed with the subsequent steps
# Ensure OUTCOME column is treated as string and handle NaN values
df_final['OUTCOME'] = df_final['OUTCOME'].fillna('Unknown').astype(str)

# Encode the winner: 1 if Fighter A wins, 0 if Fighter B wins
df_final['WINNER'] = df_final['OUTCOME'].apply(lambda x: 1 if x.startswith('W/') else 0)

# Encode the method of victory as a categorical variable
df_final['VICTORY_METHOD'] = df_final['METHOD']
```

```python
# Example: Calculate win streak for Fighter A and Fighter B
fighter_win_loss = df_final.groupby('FIGHTER_A')['WINNER'].cumsum() - df_final['WINNER']
df_final['FIGHTER_A_WIN_STREAK'] = fighter_win_loss
```
